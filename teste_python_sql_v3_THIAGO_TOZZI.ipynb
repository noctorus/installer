### 2. Criação de um novo arquivo processado

1. Formato do Arquivo: "parquet"
2. Nome "doacoes_por_partido.parquet"
3. Colunas:
    - Partido
    - Estado
    - ValorDoado - Soma de valor das doações agrupando partido e estado
    - ValorDoadoPorEstado - Soma de todas as doações agrupando somente o estado
    - Percentual - Razão entre o Valor doado para cada partido pelo valor doado Total do estado



#### Com base nas informações acima construir um codigo Python que processe os dados do arquivo "receitas_candidatos_2014_brasil_english.csv" e gere um novo arquivo somente com as colunas desejadas.

```python
# Desenvolva a Resposta Aqui
# !pip install pyarrow
# !pip install pandas
import pandas as pd
import pyarrow.parquet as pq

df_raw = pd.read_csv('receitas_candidatos_2014_brasil_english.csv') 
df_filter_partido_estado = df_raw[['cat_party','cat_election_state','num_donation_ammount']].rename(
    columns={'cat_party':'Partido','cat_election_state' : 'Estado'}).reset_index()
df_filter_estado = df_raw[['cat_election_state','num_donation_ammount']].rename(
    columns={'cat_election_state' : 'Estado'}).reset_index()
df = df_filter_partido_estado.groupby(['Partido','Estado']). \
num_donation_ammount.sum().reset_index(name='ValorDoado')
df2 = df_filter_estado.groupby(['Estado']).num_donation_ammount.sum().reset_index(name='ValorDoadoPorEstado')
df_merged = pd.merge(df, df2, on='Estado')
df_merged['Percentual'] = df_merged['ValorDoado'] / df_merged['ValorDoadoPorEstado']
print(df_merged)
df_merged.to_parquet('doacoes_por_partido.parquet')
print('arquivo salvo')
```

### 4. Lógica de programação 

1. Desenvolva utilizando **estruturas de repetição**, uma função que receba um vetor e um inteiro por parâmetro, e realize um shift no vetor do tamanho do número inteiro recebido.
        
- O tamanho do shift deverá aceitar números negativos também. O sinal do tamanho servirá para definir a direção do shift.
    - Números positivos indicam o shift da direita para a esquerda.
    - Números negativos indicam o shift da esquerda para a direita.
    
    
- **Não serão aceitas respostas do tipo:**
    - "return vetor[-tamanho:] + vetor[:-tamanho]"

```python
# Desenvolva a Resposta do item 4 aqui
def shift(vetor, tamanho):
    # tamanho = tamanho * -1
    if tamanho == 0:
        return vetor
    f = []
    t = len(vetor) - 1
    h = 1
    c = 1 
    if tamanho >= 1 : 
        i = t - tamanho
    else :
        i = (tamanho *-1) - 1
    
    while h <= len(vetor):
        i = i + c
        if i < 0:
            i = t
        elif i > t:
            i = 0
        f.append(vetor[i])
        h += 1
    
    return f
            
    
```

```python
# Cole a instrução SQL aqui
print("""
SELECT cat_candidate_name, cargo.posto, cargo.media_posto, avg(num_donation_ammount) as media_cand, cargo.media_posto / avg(num_donation_ammount)  as relacao 
FROM TABELA_DOACOES
INNER JOIN (SELECT cat_political_office as posto,
       avg(num_donation_ammount) as media_posto
FROM TABELA_DOACOES
GROUP BY 1) AS cargo on cargo.posto = TABELA_DOACOES.cat_political_office
GROUP BY 1
ORDER BY 5
""")
```
